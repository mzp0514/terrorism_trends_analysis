```python
from tqdm import tqdm
```

```python
specific_word = 'Consumerism'
```

```python
filtered_result = []
count = 0

for item in tqdm(data):
    if specific_word in item['title'] or specific_word in item['summary']:
        filtered_result.append(item)
        count += 1
        
print(count)
```
